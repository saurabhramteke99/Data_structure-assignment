Que 8.  What is a hash table, and how does it relate to dictionaries in Python?

Ans:A hash table is a data structure that allows for efficient data retrieval by using a technique called hashing. It stores key-value pairs, where each key is hashed to generate an index (or hash code) that corresponds to a position in the table, where the associated value is stored. This enables near-constant-time average performance for insertions, deletions, and lookups, assuming the hash function distributes the keys uniformly across the available slots.

Que 17.  Describe the time complexity of accessing elements in a dictionary.

Ans:In Python, dictionaries are implemented using hash tables. As a result, the time complexity for accessing an element in a dictionary is generally O(1), also known as constant time.

Here's a breakdown:

When you access an element in a dictionary using a key , Python hashes the key to determine its position in memory.
If the hash table is well-distributed (i.e., minimal collisions between keys), the lookup operation will take constant time, O(1).


However, in the case of hash collisions (when two keys hash to the same position), the time complexity may degrade. In the worst case, if many keys hash to the same position (which is highly unlikely with a good hash function and proper load factors), the time complexity could degrade to O(n), where n is the number of elements in the dictionary.

Que 19.Why are dictionaries considered unordered, and how does that affect data retrieval?

Ans:Dictionaries in Python are considered unordered because,they didn't guarantee the order of key-value pairs. The order in which items were inserted into the dictionary was not preserved when you iterated over it or retrieved items. This was because dictionaries were implemented using hash tables, which store keys in a way that optimizes retrieval but does not maintain the order of insertion.

Impact on  data retrieval.

Lookup speed: The retrieval of values based on keys remains efficient (average O(1) time complexity) because of the hash table structure, regardless of whether the dictionary maintains order.
Order preservation: In Python 3.7 and later, you can count on the insertion order being preserved when iterating through the dictionary, making it more predictable in cases where order matters.

Que 20.  Explain the difference between a list and a dictionary in terms of data retrieval?

Ans:      List:

Ordered Collection: A list stores elements in a specific order, and each element has an index (starting from 0) that determines its position in the list.
Data Retrieval: To retrieve data from a list, you access it by its index. This means you need to know the position of the element in the list.

Example.my_list = [10, 20, 30, 40]


print(my_list[2])  # Retrieves the element at index 2, which is 30


Dictionary:
Unordered Collection: A dictionary stores data as key-value pairs. The data is not stored in any particular order, and instead of indices, you retrieve values using keys.
Data Retrieval: To retrieve data from a dictionary, you use the key associated with the value. This provides a direct way to access the value without needing to know its position in the collection.
Example:


my_dict = {'a': 10, 'b': 20, 'c': 30}
print(my_dict['b'])  # Retrieves the value associated with the key 'b', which is 20
